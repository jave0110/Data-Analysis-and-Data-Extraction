```python
#########################COMPLEX Word Count
```

```python


from collections import  Counter

def complex_word_count(x):
    
    syllable = 'aeiou'
    
    t = x.split()
    
    v = []
    
    for i in t:
        words = i.split()
        c=Counter()
        
        for word in words:
            c.update(set(word))

        n = 0
        for a in c.most_common():
            if a[0] in syllable:
                if a[1] >= 2:
                    n += 1
                
        m = 0
        p = []
        for a in c.most_common():
            if a[0] in syllable:
                p.append(a[0])
        if len(p) >= 2:
            m += 1
        
        if n >= 1 or m >= 1:
            v.append(i)
            
    return len(v) 

g = h['Text'][1]

complex_word_count(g)
```

```python
df['complex_count'] = np.nan

df['complex_count'] = df['Text'].apply(lambda x: complex_word_count(x))
df.head()
```

```python
df['sentence length'] = np.nan
df['Average Sentence Length'] = np.nan
df['Percentage of Complex words'] = np.nan
df['Fog Index'] = np.nan


for i in range(0,len(df)):
    
    df['sentence length'][i]  =   len(nltk.sent_tokenize(df['Text'][i]))
    df['Average Sentence Length'][i] = df['words_counts'][i]/df['sentence length'][i]
    df['Percentage of Complex words'][i] = df['complex_count'][i]/df['words_counts'][i] 
    df['Fog Index'][i] = 0.4 * (df['Average Sentence Length'][i] + df['Percentage of Complex words'][i])
```

```python
#################################Positive Score
```

```python

def positive(x):
    
    s = x.split()
    
    positive = 0
    
    for i in dfs['Positive']:
        if i in s:
            if dfs[dfs['Positive']==i].Positive.any() == True:
                positive += 1
            
    return positive
```

```python
dfs['word_lower'] = np.nan
```

```python
df['positive_score'] = np.nan

for i in range(len(df)):
    df['positive_score'][i] = positive(df['Text'][i])
```

```python
############################NEGATIVE Score
```

```python
def negative_score(x):
    
    s = x.split()
    
    negative = 0
    
    for i in dfs['Negative']:
        if i in s:
            if dfs[dfs['Negative']==i].Negative.any() == True:
                negative += 1
            
    return negative
```

```python
df['negative_score'] = np.nan

for i in range(len(df)):
    df['negative_score'][i] = negative_score(df['Text'][i])
```

```python
df.head()
```

```python
##############################Polarity Score
```

```python
df['Polarity Score'] = np.nan

for i in range(len(df)):
    df['Polarity Score'][i] = (df['positive_score'][i]-df['negative_score'][i])/ ((df['positive_score'][i] + df['negative_score'][i]) + 0.000001)
```
